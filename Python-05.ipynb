```python
Task-9


*   Ask 2 numbers from the users and store it in num1 and num2.
*   Ask user to press 1 for addition, 2 for substraction, 3 for multiplication 
    and 4 for division.
*   Based on number given by user do the math operation.
```

```python
num1=int (input('num1='))
num2=int (input('num2='))
print('Press 1 for addition')
print('Press 2 for subtraction')
print('Press 3 for multiplication')
print('Press 4 for division')
add = num1 + num2
sub = num1 - num2
multi = num1 * num2
div = num1 / num2
Press = int (input())
print("Press=",Press)
if (Press == 1):
  print("Addition=",add)
if (Press == 2):
  print("Subtraction=",sub)
if (Press == 3):
  print("Multiplication=",multi)
if (Press == 4):
  print("Division=",div)
```
