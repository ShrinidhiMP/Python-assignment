Task-3.1
Go through other inbuilt methods like remove(),count(),clear(),sort(),reverse()
Soln:
remove()- It is an inbuilt function that removes a given object from the list.
          It does not return any value.
          Syntax:
          list_name.remove(object)

count()- It is an inbuilt function that returns count of how many time 
         a given object occurs in list.
         syntax:
         list_name.count(object)

clear()- This removes all items from the list.
         syntax:
         list.clear()

sort()- It can be used to sort the list in both ascending and descending order.
        syntex:
        For ascending order
        list_name.sort()

        For descending order
        list_name.sort(reverse=true)

reverse()- It is an inbuilt method that reverses objects of list in place.
           Syntex:
           list_name.reverse()
