Explain Return keyword with example.
-return keyword can be used to return values from a function.
It can be used to end the execution of a function.
Statements below return keyword will not get executed.

```python

def multiply(x,y):
    z = x * y
    print(z)
    return z
    print('multiply of x & y=',z) #will not be executed
my_num = multiply(10,2)
print(my_num)
    
```

Task11.2
Ask 2 numbers from users and store it in num1 and num2
1. Ask user to press 1 for addition,2 for subtraction,3 for multiplication and 4 for division
2. create 4 seperate functions for each operation (i.e. addition,subtraction,multiplication and division)based on number given by user,call that perticular function and print the output


```python
num1=int (input('num1='))
num2=int (input('num2='))
print('Press 1 for addition')
print('Press 2 for subtraction')
print('Press 3 for multiplication')
print('Press 4 for division')
Press = int (input())
if (Press == 1):
  def addition(x,y):
    z = x + y
    return z
  my_num = addition(num1,num2)
  print("Addition of num1 & num2 is",my_num)
if (Press == 2):
  def subtraction(x,y):
    z = x - y
    return z
  my_num = subtraction(num1,num2)
  print("Subtraction of num1 & num2 is",my_num)
if (Press == 3):
  def multiplication(x,y):
    z = x * y
    return z
  my_num = multiplication(num1,num2)
  print("Multiplication of num1 & num2 is",my_num)
if (Press == 4):
  def division(x,y):
    z = x / y
    return z
  my_num = division(num1,num2)
  print("Division of num1 & num2 is",my_num)
```
