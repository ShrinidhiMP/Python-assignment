Task 10.1 = what is the uasge of continue wiith an example.

continue keyword takes the execution to the begining of the loop. In other word it skips that perticular iteration.

```python
count = 20
while(count>11):
    count = count - 1
    if(count%2 == 0):
        continue
    print(count)
```
